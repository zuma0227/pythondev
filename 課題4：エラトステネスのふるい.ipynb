```python
for text in range(1,101):
    if text < 2:
        continue
    if (text != 2) and (text % 2 == 0):
        continue
    if (text != 3) and (text % 3 == 0):
        continue
    if (text != 5) and (text % 5 == 0):
        continue
    if (text != 7) and (text % 7 == 0):
        continue
    print(text)
```
