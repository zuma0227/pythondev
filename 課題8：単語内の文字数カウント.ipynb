```python
words = []
alphabet = dict.fromkeys(list("abcdefghijklmnopqrstuvwxyz"), 0)

while True:
    word = input("英単語を入力してください")
    
    if word == "":
        break

    word = word.lower()
    words.append(word)

    for text in word:
        alphabet[text] += 1

words.sort()
print(f"入力した英単語：{words}")
print(alphabet)

for key in alphabet:
    if alphabet[key] > 0:
        print(f"{key}が{alphabet[key]}個あります。")
```
